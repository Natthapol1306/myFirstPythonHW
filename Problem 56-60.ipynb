Problem 56

```python
def natthapol(r):
    a = list(sorted(r.items()))
    return([i for e,i in a])

natthapol({'x': 1, 'y': 2, 'a': 3})  

# [3, 1, 2]
```

Problem 57

```python
def invertdict(d):
    r = {}
    for k,v in d.items():
        r[v] = k
    return r


invertdict({'x': 1, 'y': 2, 'z': 3}) 
# {1: 'x', 2: 'y', 3: 'z'}
```

Problem 59

```python
import sys
import os
path = '.'
files = []
for f in os.listdir(path):
    files.append(f)
ext = [e.split('.')[-1] for e in files]
frequency = {}
for w in ext:
    frequency[w] = frequency.get(w, 0) + 1

extcount = {k: v for k, v in sorted(frequency.items(), key=lambda item: item[1], reverse=True)}
extcount = [print(v,k) for k, v in extcount.items()]
```
