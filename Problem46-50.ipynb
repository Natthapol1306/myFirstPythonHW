Problem 46

```python
def triplets(n):
    arr = []
    for a in range (1,n):
        for b in range(1,n): 
              c = b + a
        if (((a,b,c) not in arr and (b,a,c) not in arr) and c<n):
                     arr.append((a,b,c))
        return arr
```

```python
triplets(10)
```

Problem 47

```python
def enumerate(array):
    arr = []
    for i in range(len(array)):
        arr.append((i,array[i]))
    return arr
```

```python
enumerate(["a","b","c"])
```

Problem 48

```python
def array(h,w):
    arr = []
    for i in range(h):
        arr_w = []
        for e in range(w):
            arr_w = []
            for e in range(w):
                arr_w.append(None)
            arr.append(arr_w)
        return arr
```
